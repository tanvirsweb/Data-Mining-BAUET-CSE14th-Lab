```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the dataset
X = np.array([[2, 1],
              [3, 5],
              [4, 3],
              [5, 6],
              [6, 7],
              [7, 8]])

# Step 2: Mean center the data
mean = np.mean(X, axis=0)
X_centered = X - mean

# Step 3: Compute the covariance matrix
cov_matrix = np.cov(X_centered.T)

# Step 4: Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Step 5: Sort eigenvectors by decreasing eigenvalues
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

# Step 6: Project the data onto the first principal component
pc1 = eigenvectors[:, 0]
X_pca = X_centered.dot(pc1)

# Step 7: Print results
print("Mean of original data:\n", mean)
print("Covariance matrix:\n", cov_matrix)
print("Eigenvalues:\n", eigenvalues)
print("Eigenvectors (columns):\n", eigenvectors)
print("Projected data (1D PCA result):\n", X_pca)

# Step 8: Optional - Plot original and PCA direction
plt.figure(figsize=(8, 6))
plt.scatter(X_centered[:, 0], X_centered[:, 1], label='Centered Data')
for vec, val in zip(eigenvectors.T, eigenvalues):
    plt.quiver(0, 0, vec[0]*val, vec[1]*val, angles='xy', scale_units='xy', scale=1, color='r')
plt.title('PCA - Principal Component Direction')
plt.axhline(0, color='grey', linewidth=0.5)
plt.axvline(0, color='grey', linewidth=0.5)
plt.legend()
plt.grid()
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Input the dataset
X = np.array([[2, 1],
              [3, 5],
              [4, 3],
              [5, 6],
              [6, 7],
              [7, 8]])
print("Step 1: Original Data:\n", X)

# Step 2: Compute the mean of each feature
mean_vector = np.mean(X, axis=0)
print("\nStep 2: Mean Vector:\n", mean_vector)

# Step 3: Center the data by subtracting the mean
X_centered = X - mean_vector
print("\nStep 3: Centered Data:\n", X_centered)

# Step 4: Compute the covariance matrix of the centered data
cov_matrix = np.cov(X_centered.T)
print("\nStep 4: Covariance Matrix:\n", cov_matrix)

# Step 5: Compute eigenvalues and eigenvectors of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
print("\nStep 5: Eigenvalues:\n", eigenvalues)
print("\nStep 5: Eigenvectors:\n", eigenvectors)

# Step 6: Keep only the principal component with the largest eigenvalue
max_index = np.argmax(eigenvalues)
principal_component = eigenvectors[:, max_index].reshape(-1, 1)
largest_eigenvalue = eigenvalues[max_index]

print("\nStep 6: Largest Eigenvalue:\n", largest_eigenvalue)
print("\nStep 6: Principal Component:\n", principal_component)

# Step 7: Project the centered data onto the principal component (1D transformation)
projected_data = X_centered.dot(principal_component)
print("\nStep 7: Projected Data (1D):\n", projected_data)

# Step 8 (Optional): Visualize the result
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], label='Original Data')
origin = mean_vector
pc_line = principal_component * 4  # scale for visualization
plt.quiver(*origin, pc_line[0], pc_line[1], color='r', scale=1, scale_units='xy', angles='xy', label='Principal Component')
plt.title('PCA - Principal Component Direction')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.show()
```
