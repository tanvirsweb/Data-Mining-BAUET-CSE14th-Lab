# Setting up VS Code for `.ipynb` notebooks with Python 3.13

## Prerequisites

Before starting, make sure:

- You have **Python 3.13** installed.
- You have **Visual Studio Code (VS Code)** installed.

## Step 1: Install Python 3.13

1. Visit the official Python website: [https://www.python.org/downloads/release](https://www.python.org/downloads/release)

2. Download the **Python 3.13.x** installer for your OS.

3. Run the installer and check:

   - Add Python to PATH
   - Install pip (default option)

4. After installation, confirm Python version by running in terminal:

   ```bash
   python --version
   ```

   or, if on Unix/Mac:

   ```bash
   python3 --version
   ```

## Step 2: Install VS Code

1. Download from [https://code.visualstudio.com/](https://code.visualstudio.com/)
2. Install as per your operating system.

## Step 3: Install Required VS Code Extensions

In VS Code:

1. Go to the **Extensions** view (`Ctrl + Shift + X`).
2. Search for and install:

   - **Python** (by Microsoft)
   - **Jupyter** (by Microsoft)

These enable full support for `.ipynb` notebooks.

## Step 4: Create a New Notebook

1. Open a folder in VS Code where you want to store notebooks.
2. Click `File` → `New File`.
3. Save it with the extension `.ipynb`, e.g., `test_notebook.ipynb`.

## Step 5: Select Python 3.13 Interpreter

1. Click the **kernel selector** (top-right of notebook or bottom blue bar).

2. Select your **Python 3.13** interpreter. It may appear like:

   ```
   Python 3.13.0 64-bit ('base': conda) or python3.13
   ```

3. If not visible, manually locate:

   - Press `Ctrl + Shift + P` → type and select `Python: Select Interpreter`
   - Browse or enter path to `python3.13`

## Step 6: Install Jupyter Kernel for Python 3.13

If the kernel isn't available for your Python 3.13 installation:

1. Open terminal (\`Ctrl + \`\`).

2. Run:

   ```bash
   python -m pip install jupyter ipykernel
   python -m ipykernel install --user --name=python313 --display-name "Python 3.13"
   ```

3. Now go back to the notebook and choose the newly created `"Python 3.13"` kernel.

## Step 7: Test a Cell

Try running a simple code cell like:

```python
print("Hello from Python 3.13!")
```

It should execute and return the output below the cell.

## Troubleshooting

- If you see **"Python not installed"**, make sure Python 3.13 is in your system PATH.
- If Jupyter kernel doesn’t show, ensure you’ve installed `ipykernel` in the correct Python environment.

## You're Ready!

You can now:

- Write and run `.ipynb` notebooks in VS Code
- Use Python 3.13 kernel for your analysis
- Enjoy features like variable explorer, interactive plotting, and debugging!

## 6. Comparison Operators

### Python Comparison Operators compare two values and return a True or False

- `==` compares two values to see if they are the same value.
- `!=` asks if the values on each side of the operator are **not** the same.]
- `<` and `<=` are less than and less than or equal to
- `>` and `>=` are greater than and greater than or equal to, respectively.

### Let's look at some concrete examples of using comparison operators and values that return a True or a False.

- `2 < 3` returns `True` because it's a true statement that two is less than three
- `2 <= 2` returns `True` because two is less than or equal to two
- `2 < 2` returns `False` because two is not less than two
- `4 <= 2` returns `False` because four is not less than or equal to two
- `10 < 5` returns `False` because 10 is not less than 5
- `10 > 5` returns `True` because ten is greater than 5
- `2 == 2` returns `True` because the number two is equal to the number two
- `2 == 3` returns `False` because two is not three
- `3 != 2` returns `True` because three is not equal to two.
- `4 in [1, 2, 3, 4]` returns `True` because 4 exists in the list `[1, 2, 3, 4]`
- `"b" in "banana"` returns `True` because the string "b" exists inside the string `"banana"`

If we have `x = 3` followed by `result = x > 0` on the next line, the result variable will hold `True` because 3 is greater than 0.

### Comparison Operators ask the following questions

- Are these two things equal?
- Are these two things not equal?
- Is A greather than B?
- Is A greater than or equal to B?
- Is A less than B?
- Is A less than or equal to B?
- Does A exist in a list or collection of values named B


````python
print("5 > 4 is", 5 > 4)
print("1 > 0 is", 1 > 0)
# Comparison operators give us back a True or False

print(5 == 5)

print(-4 + 4 == 0)
# False because they're not the same data type
5 == "5"
print("Hello" == "Goodbye")
print("Hello" != "Goodbye") # != means "not equal to"
# = assignment# False because they're not the same data type
5 == "5"
````

## Let's Create Some Example Functions

Imagine the exercise asks us to write a function called is_five that takes in a variable and returns True/False if that variable is the number five.
The function would look like this:

```python
def is_five(x):
    result = x == 5
    return result
```

Or we could write this a little more succinctly (since returning the result is the same as returning the expression assigned to the result variable)

```python
def is_five(x):
    return x == 5
```

The result of this function is entirely dependent on the input values we send into the function.
For example:

- `is_five(10)` will return `False`
- `is_five(-5)` returns `False` because -5 is not 5.
- `is_five("hello"`) returns `False` because the string "hello" certainly isn't the number 5.
- `is_five(5)` returns `True` because 5 == 5 evaluates to True.

Notice how this function is returning the result of a comparison operation and that the comparison operator will give us a True or a False, depending on the value contained in the input variable sent into the function.

If the exercise was to make a function named is_more_than_three, we could define the function in the following way:

```python
def is_more_than_three(x):
    return x > 3
```

Notice that we're returning the result of x > 3, which can change with different inputs.
For example:

- `is_more_than_three(10)` returns `True`
- `is_more_than_three(30)` returns `True`
- `is_more_than_three(3)` returns `False` (because 3 is not greater than 3)
- `is_more_than_three(3.025`) returns `True` because 3.023 > 3 returns True.
- `is_more_than_three(-23)` returns `False` b/c negative 23 is less than 3.

````python
# how to define our own functions
# def function_name(variable_name_for_the_input):
#      output = guts of the function
#      return output
def is_even(number):
    remainder = number % 2 # the % operator returns the remainder of integer division of the number on the left by the number on the right
    if remainder == 0: # even numbers have no remainder
        return True
    else:
        return False

# calling a function means to run it
print(is_even(2))
print(is_even(4))
print(is_even(2.3))
print(is_even(3))
````

## Exercise 5: Common Data Mining Tasks


````python
import numpy as np

# Exercise 5
data = np.random.randint(0, 100, size=(5, 3))

mean_data = np.mean(data)
median_data = np.median(data)
std_data = np.std(data, axis=None)

row_means = np.mean(data, axis=1)
column_max = np.max(data, axis=0)

# Normalize the data
normalized_data = (data - mean_data) / std_data

print("Data:")
print(data)

print("Mean of the entire data:", mean_data)
print("Median of the entire data:", median_data)
print("Standard deviation of the entire data:", std_data)

print("Row means:")
print(row_means)

print("Maximum value of each column:")
print(column_max)

print("Normalized data:")
print(normalized_data)
````

````python
min_val = np.min(data)
max_val = np.max(data)
min_max_norm = (data - min_val) / (max_val - min_val)
print("Min-Max Normalized Data:")
print(min_max_norm)
````

## Basic line plot


````python
import matplotlib.pyplot as plt

# Basic line plot
year = [1950, 1970, 1990, 2010]  # X-axis values representing years
pop = [2.519, 3.692, 5.263, 6.972]  # Y-axis values representing population in billions

plt.plot(year, pop)  # Line plot connecting the points
plt.grid(True)  # Display grid for better readability
plt.show()  # Render the plot
````

## Enhanced version with labels and custom style


````python
import matplotlib.pyplot as plt

# Data
year = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
pop = [2.519, 3.039, 3.692, 4.434, 5.263, 6.145, 6.972, 7.794]

# Plotting the data
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(year, pop, marker='+', color='blue', linestyle='-', linewidth=2, markersize=6)  # Added marker, color, linestyle, linewidth, and markersize

# Adding grid, labels, and title
plt.grid(True, linestyle='--', linewidth=0.5)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population (in billions)', fontsize=14)
plt.title('World Population Over Time', fontsize=16)

# Adding annotations for data points
for i, txt in enumerate(pop):
    plt.annotate(f'{txt}', (year[i], pop[i]), textcoords="offset points", xytext=(0,10), ha='center')
# This loop adds annotations (text labels) above each data point.
# enumerate(pop): Loops through population values along with their indices.
# annotate(f'{txt}', ...): Places the population value as a label above the point.
# xytext=(0,10): Shifts the label 10 pixels upward.
# ha='center': Horizontally centers the text above the point.
# Show plot
plt.show()
````

## Dual Axis Plot (Accuracy vs Loss over Epochs)


````python
import matplotlib.pyplot as plt

# Example data for classification accuracy and loss
epochs = list(range(1, 21))
accuracy = [0.6, 0.65, 0.7, 0.72, 0.75, 0.78, 0.8, 0.82, 0.84, 0.85, 0.87, 0.88, 0.89, 0.9, 0.91, 0.92, 0.93, 0.94, 0.95, 0.96]
loss = [0.9, 0.85, 0.8, 0.75, 0.7, 0.65, 0.6, 0.55, 0.5, 0.45, 0.4, 0.35, 0.3, 0.25, 0.2, 0.18, 0.15, 0.12, 0.1, 0.08]

# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot accuracy
color = 'tab:blue'
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy', color=color)
ax1.plot(epochs, accuracy, marker='o', color=color, linestyle='-', linewidth=2, markersize=6, label='Accuracy')
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis for the loss
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Loss', color=color)
ax2.plot(epochs, loss, marker='x', color=color, linestyle='--', linewidth=2, markersize=6, label='Loss')
ax2.tick_params(axis='y', labelcolor=color)

# Add title and grid
plt.title('Model Classification Accuracy and Loss Over Epochs', fontsize=16)
fig.tight_layout()
ax1.grid(True, linestyle='--', linewidth=0.5)

# Show the plot
plt.show()
````

## U-Shaped Curve


````python
import matplotlib.pyplot as plt
import numpy as np

# Generate data for U-shaped training and validation accuracy
epochs = np.arange(1, 51)
train_accuracy = 0.6 + 0.4 * np.abs(np.sin(np.linspace(0, 3 * np.pi, 50)))
val_accuracy = 0.55 + 0.45 * np.abs(np.sin(np.linspace(0, 3 * np.pi, 50) + np.pi / 10))

# Create a figure and axis
plt.figure(figsize=(14, 8))

# Plot training accuracy
plt.plot(epochs, train_accuracy, marker='o', color='blue', linestyle='-', linewidth=2, markersize=6, label='Training Accuracy')

# Plot validation accuracy
plt.plot(epochs, val_accuracy, marker='x', color='red', linestyle='--', linewidth=2, markersize=6, label='Validation Accuracy')

# Add labels, title, and legend
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.title('Training and Validation Accuracy Over Epochs (U-Shaped Curves)', fontsize=16)
plt.legend(loc='upper right', fontsize=12)

# Add grid
plt.grid(True, linestyle='--', linewidth=0.5)

# Show plot
plt.show()
````
