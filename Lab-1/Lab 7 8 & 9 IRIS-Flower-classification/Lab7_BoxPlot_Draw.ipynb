# Draw a box-plot form the following dataset. (Dataset will be provided)

```python
import matplotlib.pyplot as plt

import numpy as np

# Dataset
data = [22, 25, 17, 19, 33, 64, 23, 17, 20, 18]

# Calculate Q1, Q3, and IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Calculate bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

# Plotting
plt.figure(figsize=(6, 4))
plt.boxplot(data, vert=True, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Boxplot with Outlier Detection')
plt.ylabel('Values')
plt.grid(True)
plt.show()
print("The outlier is =", outliers)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the data
data = {
    'C': [90, 54, 44, 87, 64, 98, 55, 67],
    'C++': [78, 65, 65, 90, 56, 78, 66, 88],
    'Java': [67, 43, 89, 78, 54, 88, 76, 87],
    'Python': [34, 90, 45, 90, 73, 95, 81, 78]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, palette="Set2")
plt.title("Boxplot of Programming Language Scores")
plt.ylabel("Scores")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'C':    [90, 54, 44, 87, 64, 98, 55, 67],
    'C++':  [78, 65, 65, 90, 56, 78, 66, 88],
    'Java': [67, 43, 89, 78, 54, 88, 76, 87],
    'Python': [34, 90, 45, 90, 73, 95, 81, 78]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(10, 6))
plt.boxplot([df['C'], df['C++'], df['Java'], df['Python']],
            labels=['C', 'C++', 'Java', 'Python'],
            patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='blue'),
            medianprops=dict(color='red'),
            whiskerprops=dict(color='green'),
            capprops=dict(color='black'))

plt.title('Box Plot of Programming Language Scores')
plt.ylabel('Scores')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


plt.savefig('boxplot_scores.png')
#plt.close()
```
