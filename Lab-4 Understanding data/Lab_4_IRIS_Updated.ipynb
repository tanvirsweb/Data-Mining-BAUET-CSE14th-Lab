```python
import pandas as pd

# Load the dataset
# url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
url = "iris.csv"
iris = pd.read_csv(url)

# Display the first few rows
display(iris.head())
```

```python
# Calculate basic statistics
display(iris.describe())

# Group by species and calculate statistics
grouped = iris.groupby('species')
display(grouped.mean())
display(grouped.median())
display(grouped.std())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for each feature
iris.hist(bins=20, figsize=(10, 10))
plt.show()

# Box plots for each feature
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, orient="h")
plt.show()

# Scatter plots
sns.pairplot(iris, hue='species', markers=["o", "s", "D"])
plt.show()

# Pair plot
sns.pairplot(iris, hue='species')
plt.show()
```

```python
# # Compute the correlation matrix
# corr_matrix = iris.corr()

# # Visualize the correlation matrix using a heatmap
# plt.figure(figsize=(8, 6))
# sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
# plt.show()

# Compute correlation matrix using only numeric columns
corr_matrix = iris.select_dtypes(include='number').corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Numeric Features")
plt.show()

```

```python
# Check for missing values in the dataset
missing_values = iris.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
# Remove rows with missing values
iris_cleaned = iris.dropna()
print("Number of rows after removing missing values:", iris_cleaned.shape[0])
```

```python
# Impute missing values with the mean(numeric)/mode(categorical) of the column
# iris_imputed = iris.fillna(iris.mean())
# print("Dataset after imputing missing values with column mean:\n", iris_imputed.head())

import pandas as pd

# Separate numeric and categorical columns
num_cols = iris.select_dtypes(include='number').columns
cat_cols = iris.select_dtypes(exclude='number').columns

# Fill numeric columns with mean
iris[num_cols] = iris[num_cols].fillna(iris[num_cols].mean())

# Fill categorical columns with mode
for col in cat_cols:
    iris[col] = iris[col].fillna(iris[col].mode()[0])

print("Dataset after imputing missing values:\n", iris.head())
```

```python
import pandas as pd
import seaborn as sns

# Load iris dataset as DataFrame
iris = sns.load_dataset("iris")  # 'species' is categorical

# Separate numeric columns
numeric_cols = iris.select_dtypes(include='number').columns

# Impute missing values in numeric columns with mean
iris_imputed = iris.copy()
iris_imputed[numeric_cols] = iris[numeric_cols].fillna(iris[numeric_cols].mean())

# Display result
print("Dataset after imputing missing numeric values:\n", iris_imputed.head())
```
