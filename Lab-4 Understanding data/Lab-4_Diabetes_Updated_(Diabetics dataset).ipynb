## Difference Between `inplace=True` and `inplace=False` in Pandas

### ðŸ”¹ Case 1: `inplace=True`
```python
df.fillna(df.mean(), inplace=True)
````

* The **original `df` is directly updated**.
* The method **returns `None`**.
* You **cannot assign** the result to another variable.

---

### ðŸ”¹ Case 2: `inplace=False` (default)

```python
df_new = df.fillna(df.mean(), inplace=False)
```

* The **original `df` remains unchanged**.
* A **new DataFrame** with filled values is returned.
* You can store it in a **new variable** (`df_new`) or **overwrite** the old one:

```python
df = df.fillna(df.mean())   # keeps changes in df
```

---

### âœ… Practical Example

```python
# Count nulls before
print("Nulls in df before:", df.isnull().values.sum())

# inplace=False â†’ returns a new DataFrame
df_new = df.fillna(df.mean())

print("Nulls in df after (new DataFrame):", df_new.isnull().values.sum())
print("Nulls in df (original unchanged):", df.isnull().values.sum())
```

ðŸ‘‰ **Output:**

* Only `df_new` has missing values filled.
* The original `df` still contains the old nulls.
