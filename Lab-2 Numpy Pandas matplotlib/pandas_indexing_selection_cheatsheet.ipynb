# ðŸ“˜ Pandas Indexing & Selection Cheatsheet

This notebook provides a complete guide and examples for commonly used Pandas DataFrame indexing and selection techniques:

- `loc[]`, `iloc[]`
- `at[]`, `iat[]`
- `get_loc()`
- And related commands for data analysis.


```python
# âœ… Importing pandas
import pandas as pd

# âœ… Sample DataFrame for all operations
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Score': [88, 92, 95]
}
df = pd.DataFrame(data, index=[1001, 1002, 1003])
df
```

```python
# âœ… df.loc[] â€” Label-based selection
# Accessing rows and columns using labels
print(df.loc[1001])                          # Row with index label 1001
print(df.loc[1002, 'Name'])                  # Value at label 1002 and column 'Name'
print(df.loc[1001:1003, ['Age','Score']])    # Multiple rows and specific columns
print(df.loc[df['Age'] > 25])                # Rows where age is greater than 25
```

```python
# âœ… df.iloc[] â€” Integer-location based selection
# Accessing rows and columns using integer indices
print(df.iloc[0])                   # First row
print(df.iloc[1, 1])                # 2nd row, 2nd column (Age of Bob)
print(df.iloc[0:2, 0:2])            # Rows 0 and 1, Columns 0 and 1
print(df.iloc[[0, 2], [1, 2]])      # Rows 0 and 2, Columns 1 and 2
```

```python
# âœ… df.at[] â€” Fast scalar access (label-based)
# Optimized for single value access using labels
print(df.at[1002, 'Score'])         # Score of Bob (label-based access)
```

```python
# âœ… df.iat[] â€” Fast scalar access (position-based)
# Optimized for single value access using integer indices
print(df.iat[2, 2])                 # Score of Charlie (position-based)
```

```python
# âœ… df.columns.get_loc() and df.index.get_loc()
# Used to find the index position of a label

col_index = df.columns.get_loc("Score")
print("Column index for 'Score':", col_index)

row_pos = df.index.get_loc(1003)
print("Row index position for label 1003:", row_pos)
```

```python
# ðŸ” Other related indexing and selection methods
print(df.head(2))            # First 2 rows
print(df.tail(1))            # Last row
print(df.sample(2))          # Random 2 rows

print(df.set_index('Name'))  # Set 'Name' as index
print(df.reset_index())      # Reset index to default integer index

print(df['Age'])             # Access column 'Age'
print(df[['Age', 'Score']])  # Access multiple columns

print(df.iloc[:, 0])         # First column using iloc
print(df.loc[:, 'Name'])     # Column selection using loc

print(df.query('Age > 25'))  # Query with string condition
```
