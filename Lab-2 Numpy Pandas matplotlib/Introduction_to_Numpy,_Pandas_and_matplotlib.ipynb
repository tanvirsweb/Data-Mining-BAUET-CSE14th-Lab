## 2. Introduction to Pandas

### 2.1 Creating DataFrames
Pandas provides a DataFrame data structure that allows us to work with labeled data efficiently. Let's start by creating a DataFrame.


```python
import pandas as pd

# Creating a DataFrame from a dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'Age': [25, 30, 22, 28, 24],
        'Salary': [50000, 60000, 45000, 70000, 55000]}

df = pd.DataFrame(data)
print("DataFrame:")
print(df)
display(df)
```

### 2.2 Selecting Data from DataFrame
Pandas provides various methods to select data from a DataFrame.

#### 2.2.1 Selecting by Position (`.iloc`)
You can use `.iloc` to select rows and columns by their integer positions.


```python
# Selecting the first row
row_0 = df.iloc[0]
print("\nSelecting by Position (iloc):")
print("First Row:")
print(row_0)

# Selecting the value in the second row and 'Age' column
value_2nd_row_age_col = df.iloc[1, df.columns.get_loc('Age')]
print("Value in the 2nd Row and 'Age' Column:", value_2nd_row_age_col)
```

#### 2.2.2 Selecting by Label (`.loc`)
You can use `.loc` to select rows and columns by their labels.


```python
# Selecting rows with specific labels
selected_rows = df.loc[[1, 3]]
print("\nSelecting by Label (loc):")
print("Selected Rows:")
print(selected_rows)

# Selecting a specific cell by label
cell_label = df.loc[2, 'Salary']
print("Value at Row 2 and 'Salary' Column:", cell_label)
```

## Pandas data frame functions with their purpose

| Function                | Purpose / Use Case                   |
| ----------------------- | ------------------------------------ |
| `df.head(n)`            | First `n` rows                       |
| `df.tail(n)`            | Last `n` rows                        |
| `df.sample(n)`          | Random sample of `n` rows            |
| `df.set_index('col')`   | Set column as index                  |
| `df.reset_index()`      | Reset the index to default integer   |
| `df['col']` or `df.col` | Select a single column (label-based) |
| `df[['col1', 'col2']]`  | Select multiple columns              |
| `df.iloc[:, 0]`         | First column (position-based)        |
| `df.loc[:, 'Name']`     | Column selection using label         |
| `df.query('Age > 25')`  | Query using string expressions       |


<br><br><br>

| Method      | Based On      | Returns       | Use For                       | Speed    |
| ----------- | ------------- | ------------- | ----------------------------- | -------- |
| `.loc[]`    | Label         | Series / DF   | Row/column access by label    | Moderate |
| `.iloc[]`   | Integer index | Series / DF   | Row/column access by position | Moderate |
| `.at[]`     | Label         | Scalar        | Fast single value by label    | Fast     |
| `.iat[]`    | Integer index | Scalar        | Fast single value by position | Fast     |
| `get_loc()` | Label         | Integer index | Get integer index of label    | Fast     |


### 2.3 Boolean Indexing
Boolean indexing allows selecting rows based on conditions.

```python
# Boolean indexing
high_salary_employees = df[df['Salary'] > 55000]
print("\nBoolean Indexing:")
print("Employees with Salary > 55000:")
print(high_salary_employees)
```

### 2.4 Setting Values in DataFrame
You can use the `.loc` method to set values in a DataFrame.

```python
# Setting a new value in the DataFrame
df.loc[0, 'Age'] = 26
# print("\nSetting Values in DataFrame:")

print("DataFrame with Updated Age:")
print(df)

# Setting a new column with default values
df['Experience'] = [2, 5, 1, 7, 3]
print("\nAdding a New Column:")
print(df)
```

### 2.5 Exercises
1. Select the last two rows of the DataFrame `df`.
2. Create a new DataFrame containing only the rows where the 'Age' is less than 30 and the 'Salary' is greater than 50000.
3. Set the salary of 'Charlie' to 48000 in the DataFrame `df`.

```python
# ðŸ“¦ Importing necessary library
import pandas as pd

# ðŸ”§ Creating a sample DataFrame for demonstration
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Score': [88, 92, 95]
}

# ðŸ§± Setting custom row labels (index)
df = pd.DataFrame(data, index=[1001, 1002, 1003])

# ðŸ” Print the DataFrame
print("Original DataFrame:\n", df, '\n')

# ---------------------------------------------
# ðŸ§  loc[]: Label-based selection
# ---------------------------------------------
print("df.loc[1001]:\n", df.loc[1001])  # Access the row with label 1001

print("\ndf.loc[1001, 'Age']:", df.loc[1001, 'Age'])  # Specific value (row label 1001, column 'Age')

print("\ndf.loc[1001:1002, ['Age', 'Score']]:\n", df.loc[1001:1002, ['Age', 'Score']])  # Row slice and selected columns

print("\ndf.loc[df['Age'] > 25]:\n", df.loc[df['Age'] > 25])  # Boolean condition filtering

# ---------------------------------------------
# ðŸ§  iloc[]: Integer-location based selection
# ---------------------------------------------
print("\ndf.iloc[0]:\n", df.iloc[0])  # Access the first row

print("\ndf.iloc[2, 1]:", df.iloc[2, 1])  # Value at 3rd row (Charlie), 2nd column (Age)

print("\ndf.iloc[0:2, 1:3]:\n", df.iloc[0:2, 1:3])  # Rows 0-1 and columns 1-2 (exclusive slicing)

print("\ndf.iloc[[0, 2], [1, 2]]:\n", df.iloc[[0, 2], [1, 2]])  # Non-contiguous row & column selection

# ---------------------------------------------
# âš¡ at[]: Fast scalar access (label-based)
# ---------------------------------------------
print("\ndf.at[1002, 'Score']:", df.at[1002, 'Score'])  # Fast access of a single value by label

# ---------------------------------------------
# âš¡ iat[]: Fast scalar access (position-based)
# ---------------------------------------------
print("\ndf.iat[2, 2]:", df.iat[2, 2])  # Value at 3rd row, 3rd column

# ---------------------------------------------
# ðŸ”¢ get_loc(): Get column/row index from label
# ---------------------------------------------
col_index = df.columns.get_loc("Score")  # Get the column index for 'Score'
print("\nColumn index for 'Score':", col_index)

row_index = df.index.get_loc(1002)  # Get the row position for label 1002
print("Row position for label 1002:", row_index)

# ---------------------------------------------
# ðŸ§ª Other related indexing techniques
# ---------------------------------------------
print("\ndf.head(2):\n", df.head(2))  # First 2 rows

print("\ndf.tail(1):\n", df.tail(1))  # Last row

print("\ndf.sample(1):\n", df.sample(1))  # Random 1-row sample

print("\ndf['Name']:\n", df['Name'])  # Access single column (Series)

print("\ndf[['Age', 'Score']]:\n", df[['Age', 'Score']])  # Access multiple columns

print("\ndf.iloc[:, 0]:\n", df.iloc[:, 0])  # First column (all rows)

print("\ndf.loc[:, 'Name']:\n", df.loc[:, 'Name'])  # All rows, column 'Name'

print("\ndf.query('Age > 28'):\n", df.query('Age > 28'))  # Query method

# ---------------------------------------------
# ðŸ§° Resetting and Setting Index
# ---------------------------------------------
print("\nResetting index:\n", df.reset_index())  # Restore default integer index

print("\nSetting 'Name' as index:\n", df.set_index('Name'))  # Set 'Name' column as index
```

- Create a DataFrame with the following data:

| Name  | Age | Marks |
|-------|-----|-------|
| John  | 22  | 85    |
| Alice | 20  | 91    |
| Bob   | 23  | 76    |
| David | 21  | 66    |

### Tasks:
1. Display only rows where `Marks > 80`.
2. Add a new column `Grade` based on:
   - A if Marks â‰¥ 90
   - B if Marks â‰¥ 80
   - C otherwise
3. Convert all values in the `Name` column to uppercase.

### Using `.loc` and `.iloc`:
1. Select the second row and third column using `.iloc`.
2. Select rows where `Age > 21` using `.loc`.
3. Set the `Marks` of David to 75.


```python
import numpy as np
import pandas as pd

# 1. Create a 3x3 NumPy array with random integers between 10 and 99
arr = np.random.randint(10, 100, size=(3, 3))
print("3x3 Array:\n", arr)

# i. Calculate the mean, maximum, and standard deviation
print("Mean:", np.mean(arr))
print("Max:", np.max(arr))
print("Standard Deviation:", np.std(arr))
```

```python
# 2. Create a Pandas DataFrame
data = {
    "Name": ["John", "Alice", "Bob", "David"],
    "Age": [22, 20, 23, 21],
    "Marks": [85, 91, 76, 66]
}
df = pd.DataFrame(data)
print("\nOriginal DataFrame:\n", df)
```

```python
# i. Display only rows where Marks > 80
print("\nRows with Marks > 80:\n", df[df["Marks"] > 80])

# ii. Add a new column Grade
def assign_grade(marks):
    if marks >= 90:
        return 'A'
    elif marks >= 80:
        return 'B'
    else:
        return 'C'

df["Grade"] = df["Marks"].apply(assign_grade)
print("\nDataFrame with Grade:\n", df)

# iii. Replace all values in Name column to uppercase
df["Name"] = df["Name"].str.upper()
print("\nName column in uppercase:\n", df)
```

```python
# Using .loc and .iloc
# i. Select the second row and third column using .iloc
second_row_third_col = df.iloc[1, 2]
print("\nSecond row, third column using iloc:", second_row_third_col)

# ii. Select rows where Age > 21 using .loc
print("\nRows where Age > 21 using loc:\n", df.loc[df["Age"] > 21])

# iii. Set the Marks of David to 75
df.loc[df["Name"] == "DAVID", "Marks"] = 75
print("\nDataFrame after setting David's Marks to 75:\n", df)
```

## 3. Data Visualization with Matplotlib

### 3.1 Line Plots
Matplotlib is a powerful library for data visualization in Python. Let's create a simple line plot.

```python
import matplotlib.pyplot as plt

# Generating data for the plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating a line plot
plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y = sin(x)')
plt.title('Simple Line Plot')
plt.show()
```

### 3.2 Scatter Plots
Scatter plots are useful for visualizing the relationship between two variables.


```python
# Generating data for the scatter plot
x = np.random.rand(50)
y = np.random.rand(50)

# Creating a scatter plot
plt.scatter(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot')
plt.show()
```

### 3.3 Bar Plots
Bar plots are used to compare different categories.


```python
# Generating data for the bar plot
categories = ['A', 'B', 'C', 'D']
values = [15, 20, 12, 18]

# Creating a bar plot
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```
