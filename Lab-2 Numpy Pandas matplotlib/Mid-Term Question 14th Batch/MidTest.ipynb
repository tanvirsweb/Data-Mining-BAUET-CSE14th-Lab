# Step 5: Add a new column named "Students diabetes" at the last attribute


```python
# Add new column "Students diabetes" with default value NaN
df["Students diabetes"] = None

# Display the updated dataframe
print(df.head())
```

# Step 6: Handle Missing Values (Using Mean Value)


```python
# Fill missing values with column mean
df.fillna(df.mean(numeric_only=True), inplace=True)

# Display the updated dataset
print(df.isnull().sum())  # Check if any missing values remain
```

# Step 7: Extract "Age" Column and Display its Data


```python
# Extract and display the "Age" column
print(df["Age_in_year"])

```

# Step 8: Apply Any Feature Engineering Technique


```python
# Create BMI feature
df["BMI"] = df["weight"] / (df["height(m)"] ** 2)

# Display the updated dataframe
print(df.head())
```

# Step 9: Normalize Data (Any Method)


```python
from sklearn.preprocessing import MinMaxScaler

# Select numerical columns for normalization
num_cols = ["Glucose", "BloodPressure", "Insulin", "SkinThickness", "Age_in_year", "BMI"]

# Apply Min-Max Scaling
scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

# Display normalized data
print(df.head())
```

# Step 10: Data Visualization and Correlation


# 1️⃣ Data Visualization


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for Age Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df["Age_in_year"], bins=20, kde=True, color="blue")
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# Box plot for Glucose Levels
plt.figure(figsize=(6, 4))
sns.boxplot(x=df["Glucose"], color="green")
plt.title("Boxplot of Glucose Levels")
plt.xlabel("Glucose Level")
plt.show()

# Scatter plot for Weight vs. Height
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["height(m)"], y=df["weight"], hue=df["result"])
plt.title("Weight vs. Height")
plt.xlabel("Height (m)")
plt.ylabel("Weight")
plt.legend(title="Diabetes (T/F)")
plt.show()
```

# 2️⃣ Data Correlation


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```
