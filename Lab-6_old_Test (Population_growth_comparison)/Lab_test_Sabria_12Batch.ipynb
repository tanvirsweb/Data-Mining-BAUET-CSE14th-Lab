```python
import pandas as pd

data = {'Country': ['Bangladesh','Bangladesh','Bangladesh','Bangladesh','Bangladesh','Bangladesh','Bangladesh','Bangladesh','Bangladesh','Bangladesh','Bangladesh'],
       'Year': [1900,1910,1920,1930,1940,1950,1960,1970,1980,1990,2000],
       'Population': [2000000,3500000,4500000,5500000,6500000,7500000,8000000,8500000,9000000,9300000,9800000]}

df1 = pd.DataFrame(data)
print("Dataframe:")
print(df1)
df1.to_csv("Bd_pop.csv")
```

```python
import pandas as pd

data = {'Country': ['China','China','China','China','China','China','China','China','China','China','China'],
       'Year': [1900,1910,1920,1930,1940,1950,1960,1970,1980,1990,2000],
       'Population': [4000000,5500000,6500000,7500000,8000000,9000000,10000000,12000000,13000000,14000000,15000000]}

df2 = pd.DataFrame(data)
print("Dataframe:")
print(df2)

df2.to_csv("china_pop.csv")
```

```python
import matplotlib.pyplot as plt

df_combined = pd.concat([df1,df2])

plt.figure(figsize=(10,6))

for country, group_data in df_combined.groupby("Country"):
    plt.plot(group_data["Year"], group_data["Population"], marker='o', label=country)
    
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Growth Comparison of Bangladesh and China")
plt.show()
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score

actual_class = [1, 0, 0, 1, 0, 1, 0, 0, 1, 0, 1, 0]
predicted_class = [1, 1, 0, 1, 0, 1, 1, 0, 1, 0, 0, 0]

cm = confusion_matrix(actual_class, predicted_class)
print("Confusion Matrix:")
print(cm)

accuracy = accuracy_score(actual_class, predicted_class)
print("Accuracy:", accuracy)

misclassification_rate = 1 - accuracy
print("Misclassification Rate:", misclassification_rate)
```
