## Generate Dataset with random numeric & non-numeric values

### Using multiple conditions (Using apply)


`````python
df["Col2_Status"] = df["col2"].apply(
    lambda x: "High" if x > 30 else ("Medium" if x > 15 else "Low")
)
display(df.head())
`````

`````python
df["Risk"] = df.apply(lambda row: "High" if row["col4"] > 50 and row["col5"] > 200 else "Low", axis=1)
display(df.head())
`````

The name row in the function is just a variable name, but it’s conventional when using `df.apply(..., axis=1)` because `each row of the DataFrame is passed as a pandas Series`.

#### pandas Series

A **pandas Series** is a **one-dimensional labeled array** capable of holding any data type: integers, floats, strings, or Python objects.

---

##### Key Features:

- **1D array**: Like a single column of a table.  
- **Index**: Each element has a label (numeric or custom).  
- **Heterogeneous types**: Can store integers, floats, strings, or mixed types.  
- **Vectorized operations**: Supports slicing, filtering, and arithmetic operations aligned by index.

`````python

# 1. Add as last column
df_1 = df.copy()
df_1["New Column Name"] = "ContantValue/None"
print("After adding as LAST column:")
display(df_1.head())

# 2. Add as first column
df_first = df.copy()
df_first.insert(0, "New Column Name", "Unknown")
print("\nAfter adding as FIRST column:")
display(df_first.head())

# 3. Add before target column (assume target = 'Outcome')
df_before_target = df.copy()
target_col = "col3"   # change this if your target column has a different name
col_index = df_before_target.columns.get_loc(target_col)
df_before_target.insert(col_index, "New Column Name", "Unknown")
print(f"\nAfter adding BEFORE {target_col} column:")
display(df_before_target.head())

# 4. Add at random column index
df_random = df.copy()
random_index = random.randint(0, len(df_random.columns))
df_random.insert(random_index, "New Column Name", "Unknown")
print(f"\nAfter adding at RANDOM index ({random_index}):")
display(df_random.head())
`````

### ⚠️ Potential Error: Correlation on Non-Numeric Columns

- The `corr()` function in Pandas only works with **numeric values**.
- If the DataFrame contains **non-numerical columns** (e.g., strings, categories), running `df.corr()` will raise an error.

#### ✅ Solutions:

1. **Select only numerical columns**:

   ```py
      num_df = df.select_dtypes(include=['number'])
      corr_matrix = num_df.corr()
   ```

2. **Encode categorical columns** (if you want to include them in correlation):

   - Use **Label Encoding** (for ordinal categories).

   ```py
      from sklearn.preprocessing import LabelEncoder

      df[df.select_dtypes(include=['object']).columns] = df.select_dtypes(include=['object']).apply(LabelEncoder().fit_transform)
   ```

   - Use **One-Hot Encoding** (for nominal categories).

   ```py
      df_encoded = pd.get_dummies(df)
      corr_matrix = df_encoded.corr()
   ```

```python
df_encoded = pd.get_dummies(df)
````

---

### Step 1: One-Hot Encoding with `pd.get_dummies(df)`

- Converts all **categorical (non-numeric) columns** into **dummy/indicator variables**.
- Each category becomes a separate binary column (`0` or `1`).
- Example:

| Gender | Age |
| ------ | --- |
| Male   | 25  |
| Female | 30  |

becomes

| Age | Gender_Female | Gender_Male |
| --- | ------------- | ----------- |
| 25  | 0             | 1           |
| 30  | 1             | 0           |

---

### Step 2: Correlation with `df_encoded.corr()`

- Computes the **pairwise correlation matrix** between all numerical features.
- Uses **Pearson correlation** by default.
- Produces a square matrix with values in the range:

  - `+1` → Perfect positive correlation
  - `-1` → Perfect negative correlation
  - `0` → No linear correlation

---

### Why This Approach?

- Correlation only works with **numeric data**.
- `pd.get_dummies()` ensures the dataset has only numeric values.
- Then `corr()` can validly compute feature relationships.
