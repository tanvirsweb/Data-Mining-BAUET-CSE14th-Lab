### ⚠️ Potential Error: Correlation on Non-Numeric Columns

- The `corr()` function in Pandas only works with **numeric values**.  
- If the DataFrame contains **non-numerical columns** (e.g., strings, categories), running `df.corr()` will raise an error.  

#### ✅ Solutions:
1. **Select only numerical columns**:
   ```py
      num_df = df.select_dtypes(include=['number'])
      corr_matrix = num_df.corr()
   ```

2. **Encode categorical columns** (if you want to include them in correlation):

   * Use **Label Encoding** (for ordinal categories).

   ```py
      from sklearn.preprocessing import LabelEncoder

      df[df.select_dtypes(include=['object']).columns] = df.select_dtypes(include=['object']).apply(LabelEncoder().fit_transform)
   ```
   * Use **One-Hot Encoding** (for nominal categories).

   ```py
      df_encoded = pd.get_dummies(df)
      corr_matrix = df_encoded.corr()
   ```

### Step 1: One-Hot Encoding with `pd.get_dummies(df)`

* Converts all **categorical (non-numeric) columns** into **dummy/indicator variables**.
* Each category becomes a separate binary column (`0` or `1`).
* Example:

| Gender | Age |
| ------ | --- |
| Male   | 25  |
| Female | 30  |

becomes

| Age | Gender\_Female | Gender\_Male |
| --- | -------------- | ------------ |
| 25  | 0              | 1            |
| 30  | 1              | 0            |

---

### Step 2: Correlation with `df_encoded.corr()`

* Computes the **pairwise correlation matrix** between all numerical features.
* Uses **Pearson correlation** by default.
* Produces a square matrix with values in the range:

  * `+1` → Perfect positive correlation
  * `-1` → Perfect negative correlation
  * `0` → No linear correlation

---

### Why This Approach?

* Correlation only works with **numeric data**.
* `pd.get_dummies()` ensures the dataset has only numeric values.
* Then `corr()` can validly compute feature relationships.
