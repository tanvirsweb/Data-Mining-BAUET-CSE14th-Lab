# 1. Visualizing the Data 

```python
import matplotlib.pyplot as plt

# Data
x = [25, 34, 22, 27, 33, 33, 31, 22, 35, 34, 67, 54, 57, 43, 50, 57, 59, 52, 65, 47, 49, 48, 35, 33, 44, 45, 38, 43, 51, 46]
y = [59, 51, 53, 78, 59, 74, 73, 57, 69, 75, 51, 32, 40, 47, 53, 36, 35, 58, 59, 50, 25, 20, 14, 42, 20, 5, 29, 27, 85, 75]

# Scatter plot
plt.scatter(x, y, c='blue', marker='o')
plt.title('Scatter Plot of X vs Y')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

# 2. Applying K-Means Clustering

```python
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


Data = {'x': [25,34,22,27,33,33,31,22,35,34,67,54,57,43,50,57,59,52,65,47,49,48,35,33,44,45,38,43,51,46],
        'y': [59,51,53,78,59,74,73,57,69,75,51,32,40,47,53,36,35,58,59,50,25,20,14,42,20,5,29,27,85,75]
       }
  
df = DataFrame(Data,columns=['x','y'])
print (df)
  
kmeans = KMeans(n_clusters=2).fit(df)
centroids = kmeans.cluster_centers_

# print the centroids
print ("The Centroid are: ", centroids)


plt.scatter(df['x'], df['y'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```

# 2. Applying K-Means Clustering 

```python
from sklearn.cluster import KMeans
import numpy as np

# Data
data = np.array(list(zip(x, y)))

# Applying KMeans with 3 clusters
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)

# Cluster centers
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Visualize clusters
plt.scatter(x, y, c=labels, cmap='viridis', marker='o')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')  # Mark cluster centers
plt.title('K-Means Clustering (3 clusters)')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

# 3. Normalization (Min-Max Scaling)


```python
from sklearn.preprocessing import MinMaxScaler

# Data
data = np.array(list(zip(x, y)))

# Apply Min-Max normalization
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

# Display scaled data
print(scaled_data)
```

# 4. Standardization (Z-Score)


```python
from sklearn.preprocessing import StandardScaler

# Apply Z-Score normalization
scaler = StandardScaler()
z_score_data = scaler.fit_transform(data)

# Display Z-score normalized data
print(z_score_data)
```
