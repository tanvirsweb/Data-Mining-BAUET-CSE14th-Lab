# New Dataframe & Handle Missing Value

```python
import pandas as pd
import numpy as np
# Sample DataFrame with missing values
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", None],
    "Age": [25, np.nan, 30, 35, 40],
    "Score": [85, 90, np.nan, 88, 92],
    "City": ["New York", "Los Angeles", None, "Chicago", "Houston"]
}
df = pd.DataFrame(data)
print("Original DataFrame with Missing Values:")
display(df)
```

## 1️⃣ Detect Missing Values

```python
print("Missing Value Count:")
display( df.isnull().sum() )
```

## 2️⃣ Drop rows with any missing values

```python
df_drop_rows = df.dropna()
print("After Dropping Rows with Missing Values:")
display(df_drop_rows)
```

## 3️⃣ Drop columns with any missing values

```python
df_drop_cols = df.dropna(axis=1)
print("After Dropping Columns with Missing Values:")
print(df_drop_cols, "\n")
```

## Drop columns with more than 60% null values

```python
nan_percentages = df.isnull().sum()/len(df)
cols_to_drop = nan_percentages[nan_percentages>.60].index
df_cleaned = df.drop(columns=cols_to_drop)

display(df_cleaned)
```

## 4️⃣ Fill missing numerical values with the mean

```python
df_fill_mean = df.copy()
df_fill_mean["Age"] = df_fill_mean["Age"].fillna(df["Age"].mean())
df_fill_mean["Score"] = df_fill_mean["Score"].fillna(df["Score"].mean())

print("After Filling Numeric Columns with Mean:")
display(df_fill_mean)
```

## 5️⃣ Fill missing values with a specific value

```python
df_fill_value = df.fillna({"Name": "Unknown", "City": "Unknown", "Age": 0, "Score": 0})
print("After Filling with Specific Values:")
display(df_fill_value)
```

## 6️⃣ Forward Fill (propagate previous values forward)

```python
# Forward Fill
df_ffill = df.ffill()
print("After Forward Fill:")
display(df_ffill)
```

## 7️⃣ Backward Fill (propagate next values backward)

```python
# Backward Fill
df_bfill = df.bfill()
print("After Backward Fill:")
display(df_bfill)
```

## 8️⃣ Fill missing categorical values with the mode (most frequent value)

```python
df_fill_mode = df.copy()
df_fill_mode["City"] = df_fill_mode["City"].fillna(df["City"].mode()[0])
print("After Filling Categorical Column with Mode:")
display(df_fill_mode)
```
