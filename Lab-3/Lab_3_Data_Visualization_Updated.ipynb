````python
import matplotlib.pyplot as plt

# Generating data for the plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating a line plot
plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y = sin(x)')
plt.title('Simple Line Plot')
plt.grid()
plt.show()
````

### Visualize multiple marker styles in one plot for comparison.


````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 10)
y = np.sin(x)

markers = ['.', ',', 'o', 'v', '^', '<', '>', '1', '2', '3', '4', 's', 'p', '*', 'h', 'H', '+', 'x', 'D', 'd', '|', '_']

plt.figure(figsize=(10,6))

for i, m in enumerate(markers[:10]):  # Plot first 10 markers for demonstration
    plt.plot(x, y+i, marker=m, label=f'marker="{m}"')  # Offset y for clarity

plt.legend(loc='upper right')
plt.title("Example of Different Markers")
plt.show()
````

# 2. Customizing bar colors and adding value


````python
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
plt.bar(categories, values, color=['red', 'green', 'blue', 'purple'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

# Display values on each bar
for i, value in enumerate(values):
    plt.text(i, value + 0.5, str(value), ha='center', fontsize=10, fontweight='bold')

plt.show()
````

### **2. Common `numpy.random` functions**

| Function                                               | Description                                           | Example                                                          |
| ------------------------------------------------------ | ----------------------------------------------------- | ---------------------------------------------------------------- |
| `np.random.rand(d0, d1, ...)`                          | Uniform random numbers in \[0,1), shape `(d0,d1,...)` | `np.random.rand(3)`  `[0.548, 0.715, 0.602]`                    |

| `np.random.randint(low, high=None, size=None)`         | Random integers `[low, high)`                         | `np.random.randint(1, 10, 5)`  `[3, 7, 1, 5, 8]`                |
| `np.random.random(size)`                               | Uniform random numbers in \[0,1), same as `rand`      | `np.random.random(4)`  `[0.23, 0.67, 0.12, 0.89]`               |
| `np.random.choice(a, size=None, replace=True, p=None)` | Randomly choose elements from array `a`               | `np.random.choice([10,20,30], 5)`  `[20,10,10,30,20]`           |
| `np.random.seed(seed)`                                 | Sets the seed for reproducibility                     | `np.random.seed(42)`                                             |
| `np.random.shuffle(x)`                                 | Shuffle array `x` in-place                            | `arr = np.array([1,2,3]); np.random.shuffle(arr)`                |
| `np.random.permutation(x)`                             | Returns a shuffled copy of `x`                        | `np.random.permutation([1,2,3])`  `[3,1,2]`                     |
| `np.random.uniform(low=0.0, high=1.0, size=None)`      | Uniform distribution over `[low, high)`               | `np.random.uniform(5,10,3)`  `[6.2, 7.8, 9.1]`                  |
| `np.random.normal(loc=0.0, scale=1.0, size=None)`      | Normal distribution with mean=loc, std=scale          | `np.random.normal(0, 1, 5)`  `[0.12, -0.34, 1.23, 0.55, -0.87]` |
| `np.random.binomial(n, p, size)`                       | Binomial distribution                                 | `np.random.binomial(10, 0.5, 5)`  `[4, 6, 5, 7, 3]`             |

# Understanding data distribution using box plots.


````python
# !pip install seaborn
````

````python
data = {'Category': np.random.choice(['A', 'B', 'C'], 100),
        'Values': np.random.randn(100) * 10 + 50}
df = pd.DataFrame(data)
display(df.head())
sns.boxplot(x='Category', y='Values', data=df)
plt.title('Box Plot Example')
plt.show()
````

### Explanation for boxplot code

```py
data = {
    'Category': np.random.choice(['A', 'B', 'C'], 100),
    'Values': np.random.randn(100) * 10 + 50
}
```

- np.random.choice(['A', 'B', 'C'], 100)

  - Generates 100 random category labels, each being either 'A', 'B', or 'C'.
  - Example output: ['B', 'A', 'C', 'B', ...]

- np.random.randn(100)

  - Generates 100 random numbers from a standard normal distribution (mean = 0, standard deviation = 1).

- \*10+50

  - Scales the data: multiplying by 10 changes the standard deviation to 10, adding 50 shifts the mean to 50.
  - This produces values roughly centered around 50 with variability.

- The result is stored in a dictionary with two keys:

  - "Category"  categorical labels (A, B, C)

  - "Values"  numerical data

```py
df = pd.DataFrame(data)
```

Converts the dictionary into a Pandas DataFrame, which organizes the data in tabular form.

Example structure:

<pre>
Category	Values
A	47.8
C	62.1
B	49.3
</pre>

```py
sns.boxplot(x='Category', y='Values', data=df)
```

- Uses Seaborns boxplot() function to visualize the distribution of "Values" for each "Category".
- X-axis: Categories (A, B, C)
- Y-axis: Numeric values from the "Values" column
- What a box plot shows:
- Box  Interquartile Range (middle 50% of the data)
- Line inside box  Median
- Whiskers  Range of data excluding outliers
- Points outside whiskers  Outliers

````python
from sklearn.datasets import load_iris
iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)

display(df_iris.head())
display(df_iris.describe().T)

print(f"{df_iris.shape = }")
print(f"{df_iris.dtypes = }")
print(f"{iris.target = }")
````

````python
df_iris['species'] = iris.target
sns.pairplot(df_iris, hue='species')
plt.show()
````

````python
df_iris['species'].hist()
````

### **1. Adding the `species` Column**

```python
df_iris['species'] = iris.target
```

- **`iris.target`** is an array of integers (`0`, `1`, `2`) representing the species of each iris flower.

  - `0`  setosa
  - `1`  versicolor
  - `2`  virginica

- This line **adds a new column** named `"species"` to `df_iris`, so each row now has both the **features** (petal length, sepal width, etc.) and the **species label**.

### **2. Pairwise Scatterplots**

```python
sns.pairplot(df_iris, hue='species')
plt.show()
```

- **`sns.pairplot()`** (from Seaborn) creates a **grid of scatter plots** showing the relationship between **every pair** of numeric columns in the DataFrame.
- **`hue='species'`**  Colors the points based on the `"species"` column, so you can visually compare how the three species are distributed in feature space.
- **Purpose**  To quickly detect:

  - Clusters
  - Correlations
  - Overlap between species

### **3. Histogram of Species Counts**

```python
df_iris['species'].hist()
```

- Creates a **histogram** showing how many samples there are for each species label (`0`, `1`, `2`).
- Since the Iris dataset is balanced, youll see **three bars of equal height** (50 samples for each species).

### ** Summary Table**

| Code Line                          | Purpose                                     |
| ---------------------------------- | ------------------------------------------- |
| `df_iris['species'] = iris.target` | Adds species label to DataFrame             |
| `sns.pairplot(..., hue='species')` | Visualizes feature relationships by species |
| `df_iris['species'].hist()`        | Shows species distribution counts           |

````python
# Map numeric labels to species names
species_mapping = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
df_iris['species_name'] = df_iris['species'].map(species_mapping)

# Plot histogram with species names
df_iris['species_name'].value_counts().plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Number of Samples per Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
````

## Task 1 & Task 2: Data Loading, Visualization & Customization


````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Iris dataset
iris = sns.load_dataset("iris")

# Display first 5 rows
display(iris.head())
````

## Visualization 1: Count Plot (Species Distribution)


````python
plt.figure(figsize=(6,4))
sns.countplot(x="species", data=iris, hue="species", palette="pastel", legend=False)
plt.title("Iris Species Distribution", fontsize=14, fontweight='bold')
plt.xlabel("Species")
plt.ylabel("Number of Samples")
plt.show()
````

## Visualization 2: Histogram (Petal Length Distribution)


````python
plt.figure(figsize=(6,4))
sns.histplot(iris["petal_length"], bins=10, color="skyblue", edgecolor="black")
plt.title("Petal Length Distribution", fontsize=14, fontweight='bold')
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.show()
````

<pre>
bins=10  Divides the data into 10 equally spaced intervals (bars).
Choosing an Appropriate bins Value for Iris
The petal length in Iris:

Minimum  1 cm

Maximum  7 cm

Range  6 cm

There are 150 samples.

General guidelines:
Too few bins  lose detail (distribution looks blocky).

Too many bins  histogram looks noisy.
</pre>


## Visualization 3: Scatter Plot (Sepal Length vs Sepal Width)


````python
plt.figure(figsize=(6,4))
sns.scatterplot(x="sepal_length", y="sepal_width", data=iris, hue="species", style="species", palette="Set2")
plt.title("Sepal Length vs Sepal Width by Species", fontsize=14, fontweight='bold')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
````

## Visualization 4: Box Plot (Petal Length by Species)


````python
plt.figure(figsize=(6,4))
sns.boxplot(x="species", y="petal_length", data=iris, hue="species", palette="muted", legend=False)
plt.title("Petal Length Distribution by Species", fontsize=14, fontweight='bold')
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.show()

# hue="species"  Ensures the palette applies correctly to each species.
# legend=False  Removes the redundant legend, keeping the plot clean
````
